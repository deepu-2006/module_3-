mention each data point names with the help of loop

```python

import matplotlib.pyplot as plt

# Data
names = ['annu', 'rinku', 'deepu', 'Akash']
age = [15, 17, 19, 18]
plt.plot(names,age,color="r", marker="d")
plt.scatter(names, age, color='blue')
# Loop to label each data point
for i in range(len(names)):
    plt.text(names[i], age[i], names[i], fontsize=10, ha='center', va='bottom')

# Add labels and title
plt.xlabel('Names' , fontsize = 20)
plt.ylabel('Age', fontsize = 20)
plt.title("Details", fontsize = 20)
plt.show()
```

we draw two lines in a single graph.

```python
# Data for lines
x = ['raju','jaggu','bheem','kaliya']
y = [1, 8, 5, 7]
z= [11, 19, 12, 17]

# Ploting both lines
plt.plot(x, y,color='y', marker='o')
plt.plot(x, z,color='g', marker='s')

# Add labels, title
plt.xlabel('Age',fontsize = 20)
plt.ylabel('Name',fontsize = 20 )
plt.title('Data',fontsize = 20)
plt.show()
```

  add legend

```python
# Data for lines

x = ['raju','jaggu','bheem','kaliya']
y = [1, 8, 5, 7]
z= [11, 19, 12, 17]

# Ploting both lines
plt.plot(x, y,color='b', marker='o',label="age")
plt.plot(x, z,color='g', marker='s',label="age")
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Details')
#for legend we need to add label while ploting lines
plt.legend()
plt.show()
```

draw subplot for line chart

```python
# Data
x = ['raju','jaggu','bheem','kaliya']
y = [1, 8, 5, 7]
z= [11, 19, 12, 17]

# Create 2 subplots (1 row, 2 columns)
plt.subplot(1, 2, 1)
plt.plot(x, y, color='b', marker='o')
plt.title('Line Chart Y',fontsize = 20)
plt.xlabel('Name',fontsize = 20)
plt.ylabel('Age',fontsize = 20)
plt.subplot(1, 2, 2)
plt.plot(x, z, color='g', marker='s')
plt.title('Line Chart Z', fontsize = 20)
plt.xlabel('Name', fontsize = 20)
plt.ylabel('Age', fontsize = 20)
# Adjust layout
plt.tight_layout()
plt.show()

```
